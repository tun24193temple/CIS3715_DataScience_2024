# Part 2. Vector and Matrix

In this lab, we will learn how to use Python to operate vectors and matrices. 

## 1. Example

### 1.1 Load data with pandas

```python
import numpy as np
import pandas as pd

# load data with pandas
df = pd.read_csv('cars.csv')
# print(df)

# convert a pandas dataframe to a numpy array
X = df.values[:, 1:-1]
print(type(X))
print(X.shape)
print(X)
```

### 1.2 Operation for vector

```python
#Get first row and the second to last row

x1 = X[0, :]
x2 = X[-2, :]

print(x1)
print(x2)

# Addition

y = x1 + x2
print(y)
```

### 1.3 Operation for matrix

```python
# Get first 5 rows and last 5 rows 

m1 = X[0:5, :]
m2 = X[-5:, :]

print(m1)
print(m2)

# Subtraction
m = m1 - m2
print(m)
```

## 2 Task

### 2.1 Get the first column and the second to last column from X

```python
# this is your code to get two vectors 
with open('output.txt', 'w') as f:
    v1 = X[:, 0]
    v2 = X[:, -2]

    print("First Column:\n", v1,file = f)
    print("Second to Last Column:\n", v2,file = f)
```

### 2.2 Perform following operations on these two vectors and print the result

```python
with open('output.txt', 'a') as f:
    # 1. addition and subtraction
    print("Addition: {}\n".format(v1+v2),file = f)
    print("Subtraction: {}\n".format(v1-v2),file = f)

    # 2. element-wise multiplication
    print("Element-Wise Multiplication: {}\n".format(np.multiply(v1, v2)),file = f)

    # 3. inner product
    print("Inner Product: {}\n".format(np.inner(v1,v2)),file = f)

    # 4. three norms
    print("First Column Vector Norms:",file = f)
    print("\tL1-Norm: {}".format(np.linalg.norm(v1, ord = 1)),file = f)
    print("\tL2-Norm: {}".format(np.linalg.norm(v1, ord = 2)),file = f)
    print("\tInf-Norm: {}".format(np.linalg.norm(v1, ord=np.inf)),file = f)
    print("Second to Last Column Vector Norms:",file = f)
    print("\tL1-Norm: {}".format(np.linalg.norm(v2, ord = 1)),file = f)
    print("\tL2-Norm: {}".format(np.linalg.norm(v2, ord = 2)),file = f)
    print("\tInf-Norm: {}".format(np.linalg.norm(v2, ord=np.inf)),file = f)
    # 5. compute the distance between these two vectors
    print("Distance: {}".format(np.linalg.norm(v1 - v2, ord = 2)),file = f)
```

### 2.3 Get the first three columns and the last three columns from X

```python
np.set_printoptions(threshold=np.inf)
with open('output.txt', 'a') as f:
    # this is your code to get two matrices: A, B
    A = X[:, 0:3]
    B = X[:, -3:]

    print("First Three Columns (Matrix A):\n", A,file = f)
    print("Last Three Columns (Matric B):\n",B,file = f)
```

### 2.4 Perform following operations on these two matrices and print the result

```python
with open('output.txt', 'a') as f:
    # 1. addition and subtraction 
    print("Addition: {}\n".format(A+B),file = f)
    print("Subtraction: {}\n".format(A-B),file = f)

    # 2. compute the transpose of the second matrix: C = B^T, and print the shape of C
    C = np.transpose(B)
    print("Second Vector Transpose: {}\n".format(C),file = f)

    # 3. compute the product between A and C, and print the result and its shape
    AC = np.dot(A,C)
    print("A*C = {}\n".format(AC),file = f)

    # 4. compute the transpose of the first matrix: D = A^T, and print the shape of D
    D = np.transpose(A)
    print("First Vector Transpose: {}\n".format(D),file = f)

    # 5. compute the product between D and B, and print the result and its shape
    BD = np.dot(B,D)
    print("B*D = {}\n".format(BD),file = f)
```
