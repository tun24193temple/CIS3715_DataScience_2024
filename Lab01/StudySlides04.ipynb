# Exploratory Data Analysis 

### EDA: Different data have different properties
- once a dataset is loaded, the first step is to learn something about it.

### Tabular Data 
- Dataset is represented as a 2d array
- rows, also reffered to as examples, data points
- columns, also reffered to as attributes, features, variables
    - Numerical Features
    - Categorical Features


```python
import pandas as pd

df = pd.read_csv('cars.csv')
df.info()
```

### Statistics of Numerical Features


```python
df.describe()
```

### Visualize Numerical Features

#### Line Plot


```python
import matplotlib.pyplot as plt

plt.plot(df['Horsepower'])
plt.show()
```

#### Hist plot: Display frequency of attribute values 

```python
df.hist(bins = 50, figsize = (20,15))
plt.show()
```

```python
plt.hist(df['MPG'], bins=50)
plt.show()
```

#### Long Tail Distribution vs Non-Long Tail Distribution
- Long Tail: Many values far from center of distribution
- Non-Long Tail: Sort of distributed

### Tabular Data

#### Boxplot: Show statistics of features
- min
- max
- media
- first quartile
- third quartile

```python
plt.boxplot(df['Acceleration'], vert = False)
plt.show()
```

### Exploring Pairs of Features
- Disclose the dependencies or correlations between the attributes

```python
plt.scatter(df['MPG'], df['Weight'])
plt.show()
```

```python
hcorr = df.corr(numeric_only = True)
hcorr.style.background_gradient()
```

### Categorical Features

```python
print(df['Origin'].value_counts())
```

### Visualize Categorical Features

```python
colors = ['blue', 'orange', 'green']
df['Origin'].value_counts().plot(kind='bar', color=colors)

plt.show()
```

```python
df['Origin'].value_counts().plot(kind = 'pie', color = colors)
plt.show()
```

### Temporal Data
- A value or values that change in time (Trending, Seasonality)

### Spatial Data
- Value or values that change over space

### Graph Data
- Dataset is represented as a graph, where nodes and edges can have their own properties and values

#### Statistics of Graph
- Nodes
- Edges
- degrees

```python
import networkx as nx

G = nx.erdos_renyi_graph(20, 0.4)

print(G.number_of_nodes())
print(G.number_of_edges())
print(G.degree([1,2]))

nx.draw(G, with_labels = True, font_weight = 'bold')
plt.show()
```

- Number of connected components

```python
G = nx.erdos_renyi_graph(20,0.1)
print(list(nx.connected_components(G)))

nx.draw(G, with_labels = True, font_weight = 'bold')
plt.show()
```

#### Visualization of graph data
